## ques13--> What is the average sale price per month (MoSold) or per year (YrSold)?

```python
# Extract month and year from 'MoSold' and 'YrSold' columns
data['MoSold'] = pd.to_datetime(data['MoSold'], format='%m').dt.month_name()
data['YearSold'] = pd.to_datetime(data['YrSold'], format='%Y')

# Group the data by month or year and calculate the average sale price
average_price_per_month = data.groupby('MoSold')['SalePrice'].mean()
average_price_per_year = data.groupby('YearSold')['SalePrice'].mean()

# Plot using line plot
plt.figure(figsize=(12, 6))

# Plot average sale price per month
plt.subplot(2, 1, 1)
average_price_per_month.plot(kind='line', marker='o', color='blue')
plt.title('Average Sale Price per Month')
plt.xlabel('Month')
plt.ylabel('Average Sale Price')

# Plot average sale price per year
plt.subplot(2, 1, 2)
average_price_per_year.plot(kind='line', marker='o', color='green')
plt.title('Average Sale Price per Year')
plt.xlabel('Year')
plt.ylabel('Average Sale Price')

plt.tight_layout()
plt.show()
```

1. Why did you pick the specific chart?

-> Line plots were chosen to visualize the average sale price per month and per year. Line plots are suitable for showing trends or patterns over time, which is the case here as you're analyzing the average sale price over different months and years. They allow for easy comparison of data points across time intervals, making them ideal for this type of analysis.Using subplots allows you to visualize multiple aspects of the data side by side, making it easier to compare and analyze them.

2. What is/are the insight(s) found from the chart?

-> The line plots illustrate the trend of average sale prices over different months and years. The chart shows the average sale price of laptop computers per month (MoSold) and per year (YrSold). The average sale price per month appears to be higher than the average sale price per year.

```python
from sklearn.model_selection import GridSearchCV

from sklearn.linear_model import Lasso
from sklearn.tree import DecisionTreeRegressor

def find_best_model_using_gridsearch(X,Y):
    algos={
        'linear_regression' : {
            'model': LinearRegression(),
            'params':{
                'positive' :[True,False]
            }
        },
        'lasso' : {
            'model': Lasso(),
            'params': {
                'alpha' : [1,2],
                'selection' :['random','cyclic']
            }
        },
        'decision_tree': {
            'model' : DecisionTreeRegressor(),
            'params': {
                'criterion' : ['squared_error','friedman_mse'],
                'splitter' : ['best','random']
            }
        }
    }
    scores=[]
    cv = ShuffleSplit(n_splits=5,test_size=0.2,random_state=10)
    for algo_name,config in algos.items():
        q= GridSearchCV(config['model'],config['params'],cv=cv, return_train_score= False)
        q.fit(X,Y)
        scores.append({
            'model' : algo_name,
            'best_score' : q.best_score_,
            'best_params' : q.best_params_
        })

    return pd.DataFrame(scores,columns=['model','best_score','best_params'])
find_best_model_using_gridsearch(X,Y)
```
